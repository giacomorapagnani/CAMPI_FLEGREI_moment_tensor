### Merge stations.xml of Stefenia Danesi into a signe file.xml, PLOT stations
## (!!! some station cannot be read !!!)

```python
import matplotlib.pyplot as plt

from obspy.clients.fdsn.client import Client
from obspy import UTCDateTime
from obspy.core.event import Catalog
from obspy.core.stream import Stream
from obspy.core.event import Event
from obspy.core.event import Origin
from obspy.core.event import Magnitude
from obspy.core.inventory import Inventory
from obspy import read_events
from obspy import read_inventory
import os
```

```python
workdir='/Users/giaco/UNI/PhD_CODE/Campi_Flegrei(stefania_danesi)'
invdir =  os.path.join(workdir,'RESP_FILES')
new_invdir=os.path.join(workdir,'RESP_FILES_new')
```

### load all stations.xml and merge them together

```python
stations_not_read=[]
inv_all=Inventory()

for fn in os.listdir(invdir):
    if fn.endswith(".xml"):
        inv_name_path= os.path.join(invdir,fn)
        try:
            inv_f=read_inventory(inv_name_path)       #read
            inv_all.extend(inv_f)


            inv_name_path_NEW=os.path.join(new_invdir,fn)
            inv_f.write(inv_name_path_NEW,format='STATIONXML')      #save
        except:
            print('!!!station:' ,fn ,'not read!!!')
            stations_not_read.append(fn)

inv_all.write(workdir + '/ALL_OV_stations_working.xml',format='STATIONXML')      #save
```
